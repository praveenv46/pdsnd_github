```python
import time
import pandas as pd
import numpy as np

# The follwing file adderss path is specific to each working syste. Therefore please change it to your system file address.

CITY_DATA = {'chicago': 'chicago.csv',
              'new york city': 'new_york_city.csv',
             'washington': 'washington.csv' } 

def get_data_frame():
    """
    Asks user to specify a city, month, and day to analyze.

    Calculated fields:
        (str) city - name of the city to analyze
        (str) month - name of the month to filter by, or "all" to apply no month filter
        (str) day - name of the day of week to filter by, or "all" to apply no day filter
        
     Returns:
        (DataFrame) df - Filtered based on the user input
        
    """
    print('Hello! Let\'s explore some US bikeshare data!')
    
    # get user input for city (chicago, new york city, washington).
    shark=0
    while shark ==0:
        count1=0
        city_name=input('\nEnter a city name\n')
        for i in CITY_DATA:
            count1=count1+1
            if i==city_name.lower(): 
                df = pd.read_csv(CITY_DATA[i])
                shark=1
            elif count1==3 and shark!=1:
                print('\nEnter a valid city name, either Chicago, Washington or New york city')
                
                
    # Creating a Month & Weekday column to from Start Time column
    df['Start Time'] = pd.to_datetime(df['Start Time'])
    df['month'] = df['Start Time'].dt.month
    df['day_of_week'] = df['Start Time'].dt.weekday
    df['hour'] = df['Start Time'].dt.hour



    # get user input for month (all, january, february, ... , june)
    Month_filter=input('\nDo you want to choose any specific month? Yes or No\n')
    if Month_filter.lower() == 'yes':
        kite=0
        g=0
        while kite==0:
            count2=0
            months = ['january', 'february', 'march', 'april', 'may', 'june']
            Big=input('\nWhich month? January, Febraury, March, April, May or June\n')
            c=Big.lower()
            for i in months:
                count2=count2+1
                if i == c:
                    g=months.index(c) + 1
                    df= df[df['month']==g]
                    kite=1
                elif count2==6 and kite!=1:
                    print('\nEnter a valid Month name')

    # get user input for day of week (all, monday, tuesday, ... sunday)
    weekday_filter=input('\nDo you want to select any specific weekday ? Yes or no\n')
    if weekday_filter.lower()=='yes':     
        knight=0
        f=0
        while knight==0:
            count3=0
            weekday = ['monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'saturday', 'sunday']
            Big1=input('\nWhich day? Monday, Tuesday, Wednesday, Thursday, Friday, Saturday or Sunday\n')
            d=Big1.lower()
            for i in weekday:
                count3=count3+1
                if i == d:
                    f=weekday.index(d)
                    df= df[df['day_of_week']==f]
                    knight=1
                elif count3==7 and knight!=1:
                    print('\nEnter valid day name')
    return df

def time_stats(df):
    """Displays statistics on the most frequent times of travel."""

    print('\nCalculating The Most Frequent Times of Travel...\n')
    start_time = time.time()

    # display the most common month
    popular_month=df['month'].value_counts().keys().tolist()
    popular_month_count=df['month'].value_counts().tolist()
    print('\nThe most common month of travel is {} and the count is {}'.format(popular_month[0], popular_month_count[0]))

    # display the most common day of week
    popular_week=df['day_of_week'].value_counts().keys().tolist()
    popular_week_count=df['day_of_week'].value_counts().tolist()
    print('\nThe most common day of the week is {} and the count is {}'.format(popular_week[0], popular_week_count[0]))

    # display the most common start hour
    popular_hour=df['hour'].value_counts().keys().tolist()
    popular_hour_count=df['hour'].value_counts().tolist()
    print('\nThe most common start hour is at {} o clock and the count is {}'.format(popular_hour[0], popular_hour_count[0]))


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def station_stats(df):
    """Displays statistics on the most popular stations and trip."""

    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # display most commonly used start station
    popular_start_station=df['Start Station'].value_counts().keys().tolist()
    popular_startstation_count=df['Start Station'].value_counts().tolist()
    print('\nThe most common Start Station is {} and the count is {}'.format(popular_start_station[0], popular_startstation_count[0]))
    
    # display most commonly used end station
    popular_end_station=df['End Station'].value_counts().keys().tolist()
    popular_endstation_count=df['End Station'].value_counts().tolist()
    print('\nThe most common End Station is {} and the count is {}'.format(popular_end_station[0], popular_endstation_count[0]))
    

    # display most frequent combination of start station and end station trip
    popular_combination=df.groupby(['Start Station', 'End Station']).size().idxmax()
    print('\nThe most common combination Station is {}'.format(popular_combination))
    


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def trip_duration_stats(df):
    """Displays statistics on the total and average trip duration."""

    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # display total travel time
    total_travel=df['Trip Duration'].sum()
    minutes_total=round(total_travel/60)
    print('\nThe total time travelled (rounded value)in minutes is {}'.format(minutes_total))


    # display mean travel time
    mean_travel=df['Trip Duration'].mean()
    minutes_mean=round(mean_travel/60)
    print('\nThe mean of travel duration (rounded value) in minutes is {}'.format(minutes_mean))


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)


def user_stats(df):
    """Displays statistics on bikeshare users."""

    print('\nCalculating User Stats...\n')
    start_time = time.time()
    #Checking if the dataset has Gender and Birth Year data
    if 'Gender' and 'Birth Year' in df.columns:

        # Display counts of user types
        counting_user=df.groupby(['User Type']).size()
        print('\nThe following table shows the list of user and count {}'.format(counting_user))


        # Display counts of gender
        counting_gender=df.groupby(['Gender']).size()
        print('\nThe following table shows the gender and count {}'.format(counting_gender))


        # Display earliest, most recent, and most common year of birth
        earliest_birth= df['Birth Year'].min()
        print('\nThe ealiest year of birth is {}'.format(earliest_birth))
        recent_birth=df['Birth Year'].max()
        print('\nThe recent year of birth is {}'.format(recent_birth))
        common_birth=df['Birth Year'].mode()[0]
        print('\nThe common year of birth is {}'.format(common_birth))
    else:
        print('Sorry no data exist')


    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

def raw_data(df):
    """Displays raw data based on user preference."""
    flag=0
    running_count=0
    while flag==0:
        raw_data=input('\nDo you want to see raw data?\n')
        if raw_data.lower()=='yes':
            running_count=running_count+1
            heading= 5*running_count
            print(df.head(heading))
        else:
            flag=1
        
    
def main():
    while True:
        df = get_data_frame()
        time_stats(df)
        station_stats(df)
        trip_duration_stats(df)
        user_stats(df)
        raw_data(df)

        restart = input('\nWould you like to restart? Enter yes or no.\n')
        if restart.lower() != 'yes':
            break


if __name__ == "__main__":
	main()

```
